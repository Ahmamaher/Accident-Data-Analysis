# **Data Cleaning**

### Dealing with incorrect data types


**Define:**
The Index column must be an integer (int) but is an object

**Code:**

```python
df['Index'] = df['Index'].str.extract('(\d+)').astype(int)
```

# Test:

```python
df.dtypes
```

**Define:**
2-The Accident Date column is an object instead of a datetime


**Code**:

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'], format='%d-%m-%Y')
```

**Test:**

```python
df.dtypes
```

**define:** Number_of_Casualties and number_of_vehicles their type Float64 but they are correct numbers

```python
df['Number_of_Casualties'] = df['Number_of_Casualties'].astype(int)
df['Number_of_Vehicles'] = df['Number_of_Vehicles'].astype(int)
```

### **Missing values**

**Define:**
1-Fill out the lost values ​​with the most common values ​​(Mode) of the classified columns

2-Fill in the lost values ​​in the arithmetic middle of the digital columns


**Code:**

```python
df['Road_Surface_Conditions'].fillna(df['Road_Surface_Conditions'].mode()[0], inplace=True)
df['Road_Type'].fillna(df['Road_Type'].mode()[0], inplace=True)
df['Weather_Conditions'].fillna(df['Weather_Conditions'].mode()[0], inplace=True)
```

**Code**:

```python
df['Number_of_Casualties'].fillna(df['Number_of_Casualties'].mean(), inplace=True)
df['Number_of_Vehicles'].fillna(df['Number_of_Vehicles'].mean(), inplace=True)
```
